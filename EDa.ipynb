```python
# function to plot any activity with a variable

def plot_variable(variable_name:str):
    sns.set_palette('pastel',desat=0.80)
    facetgrid  = sns.FacetGrid(train,hue = 'ActivityName', height= 6,aspect=2)
    facetgrid.map(sns.distplot,variable_name, hist = False).add_legend()
    plt.annotate("Stationary Activities", xy=(-0.956,17), xytext=(-0.9, 23), size=20,\
            va='center', ha='left',\
            arrowprops=dict(arrowstyle="simple",connectionstyle="arc3,rad=0.1"))
    plt.annotate("Moving Activities", xy=(0,3), xytext=(0.2, 9), size=20,\
            va='center', ha='left',\
            arrowprops=dict(arrowstyle="simple",connectionstyle="arc3,rad=0.1"))
    plt.title('activity vs {}'.format(variable_name),fontsize=12)
    plt.show()
```

# Using t-sne to reduce the Dimentions of the dataset

```python
import sklearn
from sklearn.manifold import TSNE
```

```python
# we are defining a function that would help us to apply tsne and also plot the dimentions
def apply_tsne_plot(xdata,ydata,perplexities, n_iter = 1000, img_name_prefix='t-sne'):
    for index, perplexity in enumerate(perplexities):
        # perform tsne
        print('applying t-sne with perplexity {} and {} iterations'.format(perplexity,n_iter))
        X_reduced = TSNE(verbose=2,perplexity=perplexity).fit_transform(xdata)
        print('completed....')
        # preparing data for seaborn
        print('creating plot for tsne')
        df = pd.DataFrame({'x':X_reduced[:,0],'y':X_reduced[:,1], 'label':ydata})
        
        sns.lmplot(data=df,x='x',y='y', hue='label', fit_reg=False,height=8, palette='Set1',
                  markers=['^','v','s','o','1','2'])
        plt.title('perplexity :{} and max iteration {}'.format(perplexity,n_iter))
        img_name = img_name_prefix + '_per{}_iter_{}.jpeg'.format(perplexity,n_iter)
        print('saving file in the directory....')
        plt.savefig(img_name)
        plt.show()
        print('Done...')
```

```python
X_pre_tsne = train.drop(['subject', 'Activity','ActivityName'], axis=1)
y_pre_tsne = train['ActivityName']
```

```python
apply_tsne_plot(X_pre_tsne,y_pre_tsne,perplexities=[2,5,10,20,50])
```
